```python
# ðŸŒ« Bienvenida
print("ðŸŒ« Bienvenido al sistema de monitoreo bÃ¡sico de calidad del aire")

# ðŸ“¥ Captura de datos
temperaturas = []
co2_niveles = []
pm25_niveles = []
humedades = []

print("\nIngrese 3 muestras de datos:")

for i in range(3):
    print(f"\nðŸ”¹ Muestra {i+1}")
    temp = float(input("Temperatura (Â°C): "))
    co2 = int(input("Nivel de COâ‚‚ (ppm): "))
    pm = float(input("PM2.5 (Âµg/m3): "))
    hum = float(input("Humedad (%): "))

    temperaturas.append(temp)
    co2_niveles.append(co2)
    pm25_niveles.append(pm)
    humedades.append(hum)

# ðŸ“Š CÃ¡lculo de promedios
temp_prom = sum(temperaturas) / len(temperaturas)
co2_prom = sum(co2_niveles) / len(co2_niveles)
pm25_prom = sum(pm25_niveles) / len(pm25_niveles)
hum_prom = sum(humedades) / len(humedades)

print("\nðŸ“Š Promedios:")
print(f"Temperatura promedio: {temp_prom:.2f} Â°C")
print(f"COâ‚‚ promedio: {co2_prom:.0f} ppm")
print(f"PM2.5 promedio: {pm25_prom:.2f} Âµg/m3")
print(f"Humedad promedio: {hum_prom:.2f} %")

# âš  EvaluaciÃ³n general
if co2_prom > 800 or pm25_prom > 50:
    print("\nâš  La calidad del aire es MALA.")
elif co2_prom > 600 or pm25_prom > 30:
    print("\nðŸŸ¡ Calidad del aire REGULAR.")
else:
    print("\nðŸŸ¢ Calidad del aire BUENA.")

# ðŸ“ˆ MÃ­nimos y mÃ¡ximos con diccionario
datos = {
    "Temperatura": temperaturas,
    "CO2": co2_niveles,
    "PM2.5": pm25_niveles,
    "Humedad": humedades
}

for parametro in datos:
    print(f"\n{parametro}:")
    print(f"  MÃ­nimo: {min(datos[parametro])}")
    print(f"  MÃ¡ximo: {max(datos[parametro])}")

# ðŸ“Š GrÃ¡fica simple
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(co2_niveles, label="COâ‚‚ (ppm)", marker='o')
plt.plot(pm25_niveles, label="PM2.5 (Âµg/m3)", marker='x')
plt.title("Monitoreo de Calidad del Aire")
plt.xlabel("Muestra")
plt.ylabel("Valor")
plt.grid(True)
plt.legend()
plt.show()

# ðŸ§  FunciÃ³n para evaluar calidad por muestra
def evaluar_calidad(co2, pm):
    if co2 > 800 or pm > 50:
        return "MALA"
    elif co2 > 600 or pm > 30:
        return "REGULAR"
    else:
        return "BUENA"

print("\nEvaluaciÃ³n por muestra:")
for i in range(3):
    calidad = evaluar_calidad(co2_niveles[i], pm25_niveles[i])
    print(f"Muestra {i+1}: COâ‚‚={co2_niveles[i]} ppm, PM2.5={pm25_niveles[i]} Âµg/m3 â†’ Calidad: {calidad}")

# ðŸŒ¡ Alerta por temperatura alta
altas = [t for t in temperaturas if t > 45]
if altas:
    print(f"\nðŸ”¥ Alerta: Se registraron temperaturas superiores a 45Â°C â†’ {altas}")

# ðŸ§¾ Despedida
print("\nâœ… Gracias por usar el sistema de monitoreo bÃ¡sico de calidad del aire.")
```
